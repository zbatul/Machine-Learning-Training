## Regression problem: Prediction of pizza demanded

```python
# import statements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, preprocessing, cross_validation
from statsmodels.compat import lzip
import statsmodels.stats.api as ss
import statsmodels.formula.api as sm
import seaborn as sns
%matplotlib inline
```

```python
# read file and preprocessing
file_reg = "Ex06_Pizzademand_Regreesion.csv"
data = pd.read_csv(file_reg)
```

```python
x_reg = data.drop("PizzaDemand", axis=1)
y_reg = data.PizzaDemand
```

```python
data.head()
```

## Plot of each feature with the target to see relationship

```python
plt.scatter(data.PizzaPrice, y_reg)
plt.xlabel("Pizza Price")
plt.ylabel("Pizza Demand")
plt.title("Pizza Price v/s Pizza Demand")
plt.show()
```

```python
plt.scatter(data.BurgerPrice, y_reg)
plt.xlabel("Burger Price")
plt.ylabel("Pizza Demand")
plt.title("Burger Price v/s Pizza Demand")
plt.show()
```

```python
plt.scatter(data.RoomTemperature, y_reg)
plt.xlabel("Room Temperature")
plt.ylabel("Pizza Demand")
plt.title("Room Temperature v/s Pizza Demand")
plt.show()
```

```python
plt.scatter(data.DeliveryMode, y_reg)
plt.xlabel("Delivery Mode")
plt.ylabel("Pizza Demand")
plt.title("Delivery Mode v/s Pizza Demand")
plt.show()
```

```python
plt.scatter(data.Income, y_reg)
plt.xlabel("Income")
plt.ylabel("Pizza Demand")
plt.title("Income v/s Pizza Demand")
plt.show()
```

```python
plt.scatter(data.softdrinkPrice, y_reg)
plt.xlabel("Soft drink Price")
plt.ylabel("Pizza Demand")
plt.title("Soft drink Price v/s Pizza Demand")
plt.show()
```

```python
plt.scatter(data.Location, y_reg)
plt.xlabel("Location")
plt.ylabel("Pizza Demand")
plt.title("Location v/s Pizza Demand")
plt.show()
```

## 1. Linear Regression Model

## Finding model coefficients 

```python
reg = linear_model.LinearRegression()
reg = reg.fit(x_reg, y_reg)
print(reg.coef_)
print(reg.intercept_)
```

## Predict the model

```python
x_train1, x_test1, y_train1, y_test1 = cross_validation.train_test_split(x_reg, y_reg, test_size=0.33, random_state=5)
pred_train1 = reg.predict(x_train1)
pred_test1 = reg.predict(x_test1)
```

## Plot predicted and actual distance

```python
plt.scatter(y_test1, pred_test1)
plt.xlabel("Actual Demand")
plt.ylabel("Predicted Demand")
plt.title("Actual Demand v/s Predicted Demand")
plt.show()
```

```python
plt.scatter(pred_test1, pred_test1-y_test1)
plt.xlabel("Predicted Demand")
plt.ylabel("Error")
plt.title("Predicted Demand v/s Error")
plt.show()
```

MEAN SQUARE ERROR

```python
# mean square error train data
train_mse1 = np.mean((y_train1-pred_train1)**2)
# mean square error test data
test_mse1 = np.mean((y_test1-pred_test1)**2)
print(train_mse1)
print(test_mse1)
```

MEAN OF ERROR (APPROXIMATELY ZERO)

```python
(pred_test1-y_test1).mean()
```

```python
corr = data.corr()
plt.figure()
sns.heatmap(corr, square=True)
plt.show()
```

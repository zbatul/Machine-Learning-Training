## Regression problem: Prediction of pizza demanded

```python
# import statements
import pandas as pd
import math
from sklearn import linear_model, cross_validation
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error
```

```python
# Reading the Data Files
data_pizza = pd.read_csv("Ex06_Pizzademand_Regreesion.csv")
data_pizza_stan = (data_pizza-data_pizza.mean())/data_pizza.std()
```

```python
y_reg = data_pizza_stan['PizzaDemand']
x_reg = data_pizza_stan.drop(['PizzaDemand'], axis=1)
```

## Basic Analysis : Correlation Matrix

```python
f, ax = plt.subplots(figsize=(12, 8))
corr = data_pizza.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Creating the test and train datasets
x_train, x_test, y_train, y_test = cross_validation.train_test_split(x_reg, y_reg, test_size=0.33, random_state=5)
```

```python
# Normalising the Data
x_train_sc = (x_train-x_train.mean())/x_train.std()
y_train_sc = (y_train-y_train.mean())/y_train.std()
x_test_sc = (x_test-x_test.mean())/x_test.std()
```

## 1. Linear Regression Model

FINDING MODEL COEFFICIENTS

```python
reg = linear_model.LinearRegression()
reg.normalize=True
reg.fit (x_train_sc, y_train_sc)
tr_coeffs = reg.coef_
tr_resi = reg.residues_
tr_intr = reg.intercept_
print("Coeffs: ", tr_coeffs)
print("Residuals: ", tr_resi)
print("Intercept: ", tr_intr)
```

PREDICT THE MODEL

```python
pred_lin = reg.predict(x_test_sc)
pred_lin = (pred_lin*y_test.std())+y_test.mean()
```

```python
plt.scatter(y_test,pred_lin)
plt.plot(y_test,y_test)
plt.xlabel("Actual Pizza Demand")
plt.ylabel("Predicted Pizza Demand")
plt.title("Actual Pizza Demand v/s Predicted Pizza Demand")
plt.show()
```

ERROR CALCULATION

```python
errors_abs = np.abs(y_test-pred_lin)
errors = y_test-pred_lin
```

```python
errors.mean()
```

```python
plt.scatter(y_test, errors)
plt.xlabel("Predicted Pizza Demand")
plt.ylabel("Error")
plt.title("Predicted Pizza Demand v/s Error")
plt.show()
```

HISTOGRAM

```python
y = pd.DataFrame(errors)
y.plot(kind = 'hist')
plt.xlabel("Error")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

BOX PLOT

```python
y.plot(kind = 'box')
plt.title("Box Plot")
plt.show()
```

## Linear Regression Model : OLS

```python
result = sm.ols(formula="PizzaDemand ~ PizzaPrice + BurgerPrice + RoomTemperature + DeliveryMode + Income + softdrinkPrice + Location", data=data_pizza_stan[:int(0.8*len(data_pizza_stan))]).fit()
result.summary()
```

PREDICT THE MODEL

```python
pred_ols = result.predict(x_test_sc)
pred_ols = (pred_ols*y_test.std())+y_test.mean()
```

```python
errors_abs = np.abs(y_test-pred_ols)
errors = y_test-pred_ols
```

## 2. Linear Regression Model : Ridge

```python
alphas_r = np.linspace(0.0001,15,1000)*0.5
```

```python
ridge = linear_model.Ridge(normalize=True)
coeffs = []
for a in alphas_r:
    ridge.set_params(alpha=a)
    ridge.fit(np.array(x_train_sc), np.array(y_train_sc))
    coeffs.append(ridge.coef_)
ax = plt.gca()
ax.plot(2*alphas_r, coeffs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title("Ridge Regression Plot")
plt.show()
```

```python
ridgecv = linear_model.RidgeCV(alphas=alphas_r, scoring='neg_mean_squared_error', normalize=True)
ridgecv.fit(np.array(x_train_sc), np.array(y_train_sc))
ridge.set_params(alpha=ridgecv.alpha_)
ridge.fit(np.array(x_train_sc), np.array(y_train_sc))
pred = ridge.predict(x_test_sc)
mean_squared_error(y_test, (pred*y_test.std())+y_test.mean())
```

```python
print(pd.Series(ridge.coef_, index=x_test_sc.columns))
print("Intercept: ", ridge.intercept_)
print("Aplha: ", ridge.alpha)
```

PREDICT THE MODEL

```python
pred_ridge = ridge.predict(np.array(x_test_sc))
pred_ridge = (pred_ridge*y_test.std())+y_test.mean()
```

```python
plt.scatter(y_test, pred_ridge)
plt.plot(y_test, y_test)
plt.xlabel("Actual Pizza Demand")
plt.ylabel("Predicted Pizza Demand")
plt.title("Actual Pizza Demand v/s Predicted Pizza Demand")
plt.show()
```

ERROR CALCULATION

```python
errors_abs = np.abs(np.array(y_test) - pred_ridge)
errors = np.array(y_test).reshape(33,1) - pred_ridge.reshape(33,1)
```

```python
errors.mean()
```

```python
errors.var()
```

```python
plt.scatter(np.array(y_test).reshape(33,1), errors)
plt.xlabel("Predicted Pizza Demand")
plt.ylabel("Error")
plt.title("Predicted Pizza Demand v/s Error")
plt.show()
```

HISTOGRAM OF ERRORS : RIDGE

```python
y = pd.DataFrame(errors)
y.plot(kind = 'hist')
plt.xlabel("Error")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

BOX PLOT

```python
y.plot(kind = 'box')
plt.title("Box Plot")
plt.show()
```

## 3. Linear Regression Model : Lasso

```python
alphas = np.linspace(0.00001,1,1000)*0.5
```

```python
lasso = linear_model.Lasso(max_iter = 10000, normalize=True)
coeffs = []
for al in alphas:
    lasso.set_params(alpha=al)
    lasso.fit(np.array(x_train_sc), np.array(y_train_sc))
    coeffs.append(lasso.coef_)
ax = plt.gca()
ax.plot(2*alphas, coeffs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title("Lasso Regression")
plt.show()
```

```python
lassocv = linear_model.LassoCV(alphas=None, cv=10, max_iter=100000, normalize=True)
lassocv.fit(np.array(x_train_sc), np.array(y_train_sc))
lasso.set_params(alpha=lassocv.alpha_)
lasso.fit(np.array(x_train_sc), np.array(y_train_sc))
pred = lasso.predict(x_test_sc)
mean_squared_error(y_test, (pred*y_test.std())+y_test.mean())
```

```python
print(pd.Series(lasso.coef_, index=x_test_sc.columns))
print("Intercept: ", lasso.intercept_)
print("Aplha: ", lasso.alpha)
```

PREDICT THE MODEL

```python
pred_lasso = lasso.predict(x_test_sc)
pred_lasso = (pred_lasso*y_test.std())+y_test.mean()
```

```python
plt.scatter(y_test, pred_lasso)
plt.plot(y_test, y_test)
plt.xlabel("Actual Pizza Demand")
plt.ylabel("Predicted Pizza Demand")
plt.title("Actual Pizza Demand v/s Predicted Pizza Demand")
plt.show()
```

ERROR CALCULATION

```python
errors_abs = np.abs(np.array(y_test)-pred_lasso)
errors = np.array(y_test).reshape(33,1)-pred_lasso.reshape(33,1)
```

```python
errors.mean()
```

```python
plt.scatter(np.array(y_test).reshape(33,1), errors)
plt.xlabel("Predicted Pizza Demand")
plt.ylabel("Error")
plt.title("Predicted Pizza Demand v/s Error")
plt.show()
```

HISTOGRAM : LASSO

```python
y = pd.DataFrame(errors)
y.plot(kind = 'hist')
plt.xlabel("Error")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

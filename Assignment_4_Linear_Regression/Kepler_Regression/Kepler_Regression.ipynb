## Regression Problem : Predict the orbital period of the planets from the data and check if they confirm to Keplerâ€™s Third Law of Planetary Motion.

```python
# import statements
import pandas as pd
import math
from sklearn import linear_model, cross_validation
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error
```

```python
# Reading the data file and preprocessing
kepler_data = pd.read_csv("Ex01_Kepler_Regression.csv")
kepler_data['planet'] = pd.Categorical(kepler_data.planet).codes
kepler_data = kepler_data.drop('planet', axis=1)
kepler_data = kepler_data.apply(lambda x: np.log(x))
```

```python
y_reg = kepler_data['OrbitalPeriod']
x_reg = kepler_data.drop(['OrbitalPeriod'], axis=1)
```

```python
f, ax = plt.subplots(figsize=(12, 8))
corr = kepler_data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Creating the test and train datasets
x_train, x_test, y_train, y_test = cross_validation.train_test_split(x_reg, y_reg, test_size=0.33, random_state=5)
```

```python
# Standardising the data
y_train_sc = (y_train-y_train.mean())/y_train.std()
x_train_sc = (x_train-x_train.mean())/x_train.std()
x_test_sc = (x_test-x_test.mean())/x_test.std()
```

## 1. Linear Regression Model

```python
reg = linear_model.LinearRegression(normalize=True)
reg.fit (x_train_sc, y_train_sc)
tr_coeffs = reg.coef_
tr_resi = reg.residues_
tr_intr = reg.intercept_
print(pd.Series(tr_coeffs, index=x_test_sc.columns))
print("Residuals: ", tr_resi)
print("Intercept: ", tr_intr)
```

PREDICT THE MODEL

```python
pred_lin = reg.predict(x_test_sc)
pred_lin = (pred_lin*y_test.std())+y_test.mean()
```

```python
plt.scatter(np.exp(y_test), np.exp(pred_lin))
plt.plot(np.exp(y_test), np.exp(y_test))
plt.xlabel("Actual Orbital Period")
plt.ylabel("Predicted Orbital Period")
plt.title("Actual Orbital Period v/s Predicted Orbital Period")
plt.show()
```

ERROR CALCULATION

```python
errors_abs = np.abs(np.exp(y_test)-np.exp(pred_lin))
errors = np.exp(y_test)-np.exp(pred_lin)
```

```python
plt.scatter(np.exp(y_test), errors)
plt.xlabel("Predicted Orbital Period")
plt.ylabel("Error")
plt.title("Predicted Orbital Period v/s Error")
plt.show()
```

## 3. Linear Regression Model : Lasso

```python
alphas = np.linspace(0.0001,1,1000)*0.5
```

```python
lasso = linear_model.Lasso(max_iter = 10000, normalize=True)
coeffs = []
for al in alphas:
    lasso.set_params(alpha=al)
    lasso.fit(np.array(x_train_sc), np.array(y_train_sc))
    coeffs.append(lasso.coef_)
ax = plt.gca()
ax.plot(2*alphas, coeffs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title("Lasso regression")
plt.show()
```

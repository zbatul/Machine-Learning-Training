## Regression Problem : Given a carâ€™s make, model and other attributes, predict the fuel efficiency.

```python
# import statements
import pandas as pd
import math
from sklearn import linear_model, cross_validation
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error
import plotly.plotly as py
import plotly.graph_objs as go
from pandas.tools.plotting import scatter_matrix
```

```python
# Reading the data file and preprocessing
mpg_data = pd.read_csv("Ex04_Mpg_Regression.csv")
mpg_data = shuffle(mpg_data)
mpg_data.is_copy=False
```

Creating the Dataset

```python
mpg_data['horsepower'] = mpg_data['horsepower'].str.replace("?", "0.0")
mpg_data['horsepower'] = mpg_data['horsepower'].astype(float)
```

```python
mpg_data_car = list(mpg_data['carname'])
carname = []
for ind in range(len(mpg_data_car)):
    carname.append(mpg_data_car[ind].split()[0])
uniq_car = set(carname)
car_to_ix = { car:i for i,car in enumerate(uniq_car)}
mpg_data['carname'] = [car_to_ix[car.split()[0]] for car in mpg_data_car]
```

```python
y_reg = mpg_data['mpg']
x_reg = mpg_data.drop(['mpg'], axis=1)
```

## Basic Analysis : Correlation Matrix

```python
f, ax = plt.subplots(figsize=(12, 8))
corr = mpg_data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
plt.title("Correlation Matrix")
plt.show()
```

```python
scatter_matrix(mpg_data, alpha=0.2, figsize=(15,10), diagonal='kde')
plt.show()
```

```python
# Creating the test and train datasets
x_train, x_test, y_train, y_test = cross_validation.train_test_split(x_reg, y_reg, test_size=0.33, random_state=5)
```

Standardising the data

```python
mpg_data_stan = (mpg_data-mpg_data.mean())/mpg_data.std()
```

```python
x_train_sc = (x_train-x_train.mean())/x_train.std()
x_test_sc = (x_test-x_test.mean())/x_test.std()
y_train_sc = (y_train-y_train.mean())/y_train.std()
```

## 1. Linear Regression Model

FINDING MODEL COEFFICIENTS

```python
reg = linear_model.LinearRegression()
reg.normalize=True
reg.fit (x_train_sc, y_train_sc)
tr_coeffs = reg.coef_
tr_resi = reg.residues_
tr_intr = reg.intercept_
print("Coeffs: ", tr_coeffs)
print("Residuals: ", tr_resi)
print("Intercept: ", tr_intr)
```

PREDICT THE MODEL

```python
pred_lin = reg.predict(x_test_sc)
pred_lin = (pred_lin*y_test.std())+y_test.mean()
```

```python
plt.scatter(y_test, pred_lin)
plt.plot(y_test, y_test)
plt.xlabel("Actual MPG")
plt.ylabel("Predicted MPG")
plt.title("Actual MPG v/s Predicted MPG")
plt.show()
```

ERROR CALCULATION

```python
errors = pred_lin - y_test
```

```python
errors.mean()
```

```python
plt.scatter(y_test, errors)
plt.xlabel("Predicted MPG")
plt.ylabel("Error")
plt.title("Predicted MPG v/s Error")
plt.show()
```

HISTOGRAM

```python
y = pd.DataFrame(errors)
y.plot(kind = 'hist')
plt.xlabel("Error")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

BOX PLOT

```python
y.plot(kind = 'box')
plt.title("Box Plot")
plt.show()
```

## Linear Regression Model : OLS

```python
result = sm.ols(formula="mpg ~ cylinders + displacement + horsepower + weight + modelyear + origin + carname", data=mpg_data_stan).fit()
result.summary()
```

PREDICT THE MODEL

```python
pred_ols = result.predict(x_test_sc)
pred_ols = (pred_ols*y_test.std())+y_test.mean()
```

```python
plt.scatter(y_test, pred_ols)
plt.plot(y_test, y_test)
plt.xlabel("Actual MPG")
plt.ylabel("Predicted MPG")
plt.title("Actual MPG v/s Predicted MPG")
plt.show()
```

## 2. Linear Regression Model : Ridge

```python
alphas_r = np.linspace(0.0001,15,1000)*0.5
```

```python
ridge = linear_model.Ridge(normalize=True)
coeffs = []
for a in alphas_r:
    ridge.set_params(alpha=a)
    ridge.fit(np.array(x_train_sc), np.array(y_train_sc))
    coeffs.append(ridge.coef_)
ax = plt.gca()
ax.plot(2*alphas_r, coeffs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title("Ridge Regression Plot")
plt.show()
```

```python
ridgecv = linear_model.RidgeCV(alphas=alphas_r, scoring='neg_mean_squared_error', normalize=True)
ridgecv.fit(np.array(x_train_sc), np.array(y_train_sc))
ridge.set_params(alpha=ridgecv.alpha_)
ridge.fit(np.array(x_train_sc), np.array(y_train_sc))
pred = ridge.predict(x_test_sc)
mean_squared_error(y_test, (pred*y_test.std())+y_test.mean())
```

```python
print(pd.Series(ridge.coef_, index=x_test_sc.columns))
print("Intercept: ", ridge.intercept_)
print("Aplha: ", ridge.alpha)
```

PREDICT THE MODEL

```python
pred_ridge = ridge.predict(np.array(x_test_sc))
pred_ridge = (pred_ridge*y_test.std())+y_test.mean()
```

```python
plt.scatter(y_test, pred_ridge)
plt.plot(y_test, y_test)
plt.xlabel("Actual MPG")
plt.ylabel("Predicted MPG")
plt.title("Actual MPG v/s Predicted MPG")
plt.show()
```

ERROR CALCULATION

```python
errors_abs = np.abs(np.array(y_test) - pred_ridge)
errors = np.array(y_test).reshape(132,1) - pred_ridge.reshape(132,1)
```

```python
errors.mean()
```

```python
errors.var()
```

```python
plt.scatter(np.array(y_test).reshape(132,1), errors)
plt.xlabel("Predicted MPG")
plt.ylabel("Error")
plt.title("Predicted MPG v/s Error")
plt.show()
```

HISTOGRAM OF ERRORS : RIDGE

```python
y = pd.DataFrame(errors)
y.plot(kind = 'hist')
plt.xlabel("Error")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

BOX PLOT

```python
y.plot(kind = 'box')
plt.title("Box Plot")
plt.show()
```

## 3. Linear Regression Model : Lasso

```python
alphas = np.linspace(0.001,0.1,10000)*0.5
```

```python
lasso = linear_model.Lasso(max_iter = 10000, normalize=True)
coeffs = []
for al in alphas:
    lasso.set_params(alpha=al)
    lasso.fit(np.array(x_train_sc), np.array(y_train_sc))
    coeffs.append(lasso.coef_)
ax = plt.gca()
ax.plot(2*alphas, coeffs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title("Lasso Regression")
plt.show()
```

```python
lassocv = linear_model.LassoCV(alphas=None, cv=10, max_iter=100000, normalize=True)
lassocv.fit(np.array(x_train_sc), np.array(y_train_sc))
lasso.set_params(alpha=lassocv.alpha_)
lasso.fit(np.array(x_train_sc), np.array(y_train_sc))
pred = lasso.predict(x_test_sc)
mean_squared_error(y_test, (pred*y_test.std())+y_test.mean())
```

```python
print("Coeffs: ", lasso.coef_)
print("Intercept: ", lasso.intercept_)
print("Aplha: ", lasso.alpha)
```

PREDICT THE MODEL

```python
pred_lasso = lasso.predict(x_test_sc)
pred_lasso = (pred_lasso*y_test.std())+y_test.mean()
```

```python
plt.scatter(y_test, pred_lasso)
plt.plot(y_test,y_test)
plt.xlabel("Actual MPG")
plt.ylabel("Predicted MPG")
plt.title("Actual MPG v/s Predicted MPG")
plt.show()
```

ERROR CALCULATION

```python
errors_abs = np.abs(pred_lasso-np.array(y_test))
errors = pred_lasso.reshape(132,1)-np.array(y_test).reshape(132,1)
```

```python
errors.mean()
```

```python
plt.scatter(np.array(y_test).reshape(132,1), errors)
plt.show()
```

HISTOGRAM

```python
y = pd.DataFrame(errors)
y.plot(kind = 'hist')
plt.xlabel("Error")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

BOX PLOT

```python
y.plot(kind = 'box')
plt.title("Box Plot")
plt.show()
```

## Analysis of Bigrams

1. Take any corpus(article) of your choice and analyze all the bigrams present based on ConditionalFrequencyDist and compute the co-occurence matrix.
2. Visualize the co-occurence matrix.

```python
from nltk import bigrams, trigrams
from nltk.tokenize import word_tokenize
import nltk
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Take any corpus(article) of your choice and analyze all the bigrams present based on ConditionalFrequencyDist and compute the co-occurence matrix.

```python
content = word_tokenize("""The agent will book the tickets to the show for the entire family.
But you can generally book tickets online.
When you book tickets online they provide you with a book of stamps""")
```

## Visualize the co-occurence matrix for POS tags

```python
tags = nltk.pos_tag(content)
bigram_list = list(bigrams(tags))
cfd = nltk.ConditionalFreqDist((i[1], j[1]) for (i, j) in bigram_list)
df = pd.DataFrame(cfd).fillna(0).T
df
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df)
plt.title("Co-occurence matrix of POS tags of the data")
plt.show()
```

## Visualize the co-occurence matrix for word tokens

```python
bigram_list_word = list(bigrams(content))
cfd_new = nltk.ConditionalFreqDist((i, j) for (i, j) in bigram_list_word)
df_new = pd.DataFrame(cfd_new).fillna(0).T
df_new.head()
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df_new)
plt.title("Co-occurence matrix of word tokens of the data")
plt.show()
```

## Visualize the co-occurence matrix with trigrams

```python
trigrams_as_bigrams = []
trigrams_list_word = list(trigrams(content))
trigrams_as_bigrams.extend([((t[0],t[1]), t[2]) for t in trigrams_list_word])
cfd_trigram = nltk.ConditionalFreqDist((i, j) for (i, j) in trigrams_as_bigrams)
# cfd_trigram = nltk.ConditionalFreqDist([(i + ' ' + j, k) for i, j, k in trigrams_list_word])
df_tri = pd.DataFrame(cfd_trigram).fillna(0).T
df_tri
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df_tri)
plt.title("Co-occurence matrix of word tokens of the data")
plt.show()
```
